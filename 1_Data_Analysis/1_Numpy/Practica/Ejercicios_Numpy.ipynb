## Ejercicio 2
Extrae todos los impares de `my_array`:
1. Mediante bucle for
2. Mediante list comprehension
3. Mediante máscaras

````python
print(my_array)
````

````python
impares = []

for x in my_array:
    if x % 2 == 1:
        # print(x)
        impares.append(x)

print(impares)
````

````python
impares = [x for x in my_array if x % 2 == 1]
print(impares)
````

````python
my_array[my_array % 2 == 1]
````

````python
# and &
# or |
# not ~
````

````python
mask = (my_array % 2 == 1) & (my_array > 5)
my_array[mask]
````

````python
mask = (my_array % 2 == 1) | (my_array > 5)
my_array[mask]
````

````python
mask = (my_array % 2 == 1) & ~(my_array > 5)
my_array[mask]
````

## Ejercicio 15
Encuentra los índices de los valores que no sean 0 del siguiente vector [1,2,0,0,4,0]. El output sería [0, 1, 4]. Encapsúlalo en una función, con un argumento de entrada de un vector.

````python
ejer_15 = np.array([1,2,0,0,4,0])
print(ejer_15)
````

````python
np.where(ejer_15 != 0)[0]
````

````python
# ejer_15[np.where(ejer_15 != 0)[0]]
# ejer_15[[0,1,4]]
````

````python
def f_not0(array):
    return np.where(array !=0)[0]
````

````python
print(f_not0(np.array([1,6,0])))
print(f_not0(np.array([4,0,7,0])))
print(f_not0(np.array([0,0,0,0])))
print(f_not0(np.array([1,1,1,1])))
````
